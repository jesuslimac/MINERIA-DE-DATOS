```python
#Importamos las librerias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
#Importamos el conjunto de datos
store_data = pd.read_csv('groceries.csv', header=None)
```

```python
# Visualizamos el conjunto de datos
store_data.head()
```

```python
#Visializamos el las dimensiones del conjunto de datos
store_data.shape
```

```python
#Convertimos nuestro marco de datos de pandas en una lista de listas
records = []
for i in range(0, 9831):
    records.append([str(store_data.values[i,j]) for j in range(0, 41)])
```

```python
#Aplicamos el algoritmo Apriori en el conjunto de datos
association_rules = apriori(records, min_support=0.0035, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
#Encontramos el número total de reglas extraídas
print(len(association_results))
```

```python
#VIsualizamos la primera regla
print(association_results[0])
```

```python
#VIsualizamos todas las reglas devueltas
for item in association_results:
    
    # primer índice de la lista interna 
    # Contiene elemento base y elemento adicional
    pares = item[0] 
    items = [x for x in pares]
    print("Rule: " + items[0] + " -> " + items[1])
    #segundo índice de la lista interna
    print("Support: " + str(item[1]))
    #tercer índice de la lista ubicado en 0th 
    #del tercer índice de la lista interna
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
